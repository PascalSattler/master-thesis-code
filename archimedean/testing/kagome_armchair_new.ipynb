```python
# degenerate band pairs
group1 = [0,1] # upper two
group2 = [2,3] # middle two
group3 = [4,5] # lower two

# subspace EVs, is of shape (20, 20, 8, 2) -> 20 grid points, 8=2*4 bands (particle-hole), 2 EVs (eigenfunctions along band 1 and 2)
psi = EVs[:,:,:,group3]

# shift arrays to neighboring grid points (periodic boundary condition holds)
# psi_1: (k_1 + 1, k_2)
psi_1 = np.roll(psi, -1, axis=0) 
# psi_2: (k_1, k_2 + 1)
psi_2 = np.roll(psi, -1, axis=1)

# calculate link variables from multiplet states ψ
# matrix is of form M_ab = < u_a | metric | u_b >
# pre-apply metric to neighbor vectors
metric = hamil.metric

# ij for metric entries, xyjk for psi (x, y, 2*dim, band) -> xyik is metric applied to EV components (hole entries aquire minus sign)
psi_1 = np.einsum('ij,xyjk->xyik', metric, psi_1)
psi_2 = np.einsum('ij,xyjk->xyik', metric, psi_2)

# calcualte overlap U†(k) η U(k+μ)
# transpose EVs (last to dims of psi) for matrix multiplication
psi_dag = psi.conj().swapaxes(-1, -2)

# matrices
M_1 = np.matmul(psi_dag, psi_1)
M_2 = np.matmul(psi_dag, psi_2)

# determinants
U_1 = np.linalg.det(M_1)
U_2 = np.linalg.det(M_2)

# normalize to obtain link phases
U_1 /= np.abs(U_1)
U_2 /= np.abs(U_2)

# calcualte flux through plaquette U_1 * U_2(k_1 + 1) * U_1(k_2 + 2)† * U_2†
U_1_to_2 = np.roll(U_1, -1, axis=1) # U_1(k_2 + 2)
U_2_to_1 = np.roll(U_2, -1, axis=0) # U_2(k_1 + 1)

# phase product
prod = U_1 * U_2_to_1 * np.conj(U_1_to_2) * np.conj(U_2)

# principal branch of logarithm yields Berry flux
F_12 = np.imag(np.log(prod))

# sum and normalize for chern number
chern = np.sum(F_12) / (2 * np.pi)

print(chern)
```

```python
slab.parameterize_hamil(1e-4)
```

```python
dir = hamil.lat.num_rcpr_vects[0]
ks = np.linspace(0, 1, 301)
energies = np.empty((len(ks), 2*slab.lat.n_sublats))
amps = np.empty((len(ks), slab.lat.n_sublats, slab.lat.n_sublats))

for i, j in enumerate(ks):
    k = j*dir # cut dir vector into len(ks)-1 pieces
    E, EV = slab.Bogoliubov_trafo(k[0], k[1], k[2])
    amp = slab.sublat_localization(EV)
    energies[i] = E
    amps[i] = amp
```

```python
for n in range(0, slab.n_bands):
    color = amps[:, n]
    bandplot = plt.scatter(ks, energies[:,n], marker = '.', c = [(0.2298057, 0.298717966, 0.753683153, 0.8)], linewidths=0.2, edgecolor='none')
plt.tight_layout()
plt.margins(x=0, y=0)
```
