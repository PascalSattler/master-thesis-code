```python
hamil = trsqr
hamil.assume_real_parameters()
```

```python
Hmat = hamil.HP_trafo(symbolic=True)#[0:hamil.lat.n_sublats, 0:hamil.lat.n_sublats]

display(Math(r'\Large ' + latex(
    Hmat
    )))
```

```python
sym_map = {s.name: s for s in Hmat.free_symbols}
print(Hmat.free_symbols)
```
